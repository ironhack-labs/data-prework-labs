```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
import random

print("\n Welcome to the game! \n")

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
the_three_options = ['rock','paper','scissors']
results = []

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
rounds = int(input("Ok tell me, how many wins are needed to decide the winner? \n"))

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games


# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.
def func_random():
    x = random.choice(the_three_options)
    return x

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.
def your_choice():
    a = True
    while a:
        player_choice = input("Enter your choice: ").lower()
        if player_choice in the_three_options:
            a = False
        else:
            print("Oops, not valid choice. Try again.")
    return player_choice

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
def combat():
    player = your_choice()
    machine_choice = func_random()
    #print("Machine choose:",machine_choice)
    if (player == machine_choice):
        #print("Tie!")
        result = 0
    elif player == "rock":
        if machine_choice == "paper":
            #print("You lose!", machine_choice, "covers", player)
            result = 1
        else:
            #print("You win!", player, "smashes", machine_choice)
            result = 2
    elif player == "paper":
        if machine_choice == "scissors":
            #print("You lose!", machine_choice, "cut", player)
            result = 1
        else:
            #print("You win!", player, "covers", machine_choice)
            result = 2
    elif player == "scissors":
        if machine_choice == "rock":
            #print("You lose...", machine_choice, "smashes", player)
            result = 1
        else:
            #print("You win!", player, "cut", machine_choice)
            result = 2
    return player, machine_choice, result

# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
def show(player, machine_choice, winner):
    wins = results.count(2)
    loses = results.count(1)
    ties = results.count(0)
    print("\n")
    print("=="*22)
    print("Your choice is {}, the machine's choice is {}\n >>> {}!".format(player, machine_choice, winner))
    print("Up untill this moment, you've won {} times, to machine's {}. There were {} ties.".format(wins, loses, ties))
    print("\n")
    
# Create two variables that accumulate the wins of each participant


# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.


x = 0
while x < rounds:
    player, machine_choice, result = combat()
    results.append(result)
    if result == 0:
        winner = "It's a tie"
    elif result == 1:
        winner = "Machine won"
    else:
        winner = "Player won"
    wins = results.count(2)
    loses = results.count(1)
    ties = results.count(0)

    show(player, machine_choice, winner)

    x = max(wins, loses)
if wins < loses:
    print("Sorry, you lost. Goog luck next time.")
else:
    print("Congradulations, you've won!")
print("\nFinal results:\n           You've won {} times, and machine did it {} times. There were {} ties".format(wins, loses, ties))

    
# Print by console the winner of the game based on who has more accumulated wins
```
