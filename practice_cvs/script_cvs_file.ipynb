```python
data = []

with open("/Users/abreualberto91/Downloads/weight_height.csv", "r") as f:
    lines = f.readlines()
    for line in lines:
        data.append(line.split()[0].split(","))
```

```python
#obtain the avg of the actual heights for all individuals
heights = []

for person in data[1:]:
    height = int(person[2])
    heights.append(height)
```

```python
avg_height = sum(heights)/len(heights)
print(avg_height)
```

```python
#Compare females vs males
male_heights = []
female_heights =[]

for person in data[1:]:
    height = int(person[2])
    if person[0] == 'M':
        male_heights.append(height)
    elif person[0] == 'F':
        female_heights.append(height)
        
avg_male_height = sum(male_heights)/len(male_heights)
avg_female_height = sum(female_heights)/len(female_heights)

print("Average male height:", avg_male_height)
print("Average female height:", avg_female_height)
```
